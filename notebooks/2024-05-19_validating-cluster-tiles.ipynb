## Real Code

```python
grouped_by_tile = clusters.groupby("tile").count()
grouped_by_tile
```

```python
multi_ward_tiles = grouped_by_tile[grouped_by_tile["year"] > 1].reset_index()
multi_ward_tiles
```

```python
number_tiles_dropped = len(multi_ward_tiles)
number_tiles_dropped
```

```python
number_rows_dropped = multi_ward_tiles["year"].sum()
number_rows_dropped
```

```python
len(clusters)
```

```python
remaining_rows = len(clusters) - number_rows_dropped
remaining_rows
```

```python
dropped_data = clusters[clusters["tile"].isin(multi_ward_tiles["tile"])] 
dropped_data
```

```python
data_without_overlap = clusters[~clusters["tile"].isin(multi_ward_tiles["tile"])] 
data_without_overlap
```
