# Conversion of Geotiff to PNG


```python
import rasterio as rio
from rasterio.plot import show, show_hist
```

```python
image_file = '../data/images/trial-image/composite.tif'

# Open the GeoTiff file
with rio.open(image_file) as src:
    print(src.meta)
    show(src)
```

```python

with rio.open(image_file) as src:
    show_hist(src, bins=50)
```

```python
with rio.open(image_file) as src:
    arr = src.read([6,4,2])
    print(arr.min())
    print(arr.max())
    # Convert the 16-bit data to 8-bit by scaling it to the range 0-255
    # arr[0] = (arr[0] / np.amax(arr[0]) * 255).astype('uint8')
    # arr[1] = (arr[1] / np.amax(arr[1]) * 255).astype('uint8')
    # arr[2] = (arr[2] / np.amax(arr[2]) * 255).astype('uint8')
    scaled_data = (arr / (2**16 - 1) * 255).astype('uint8')
    # scaled_data = arr
    print(scaled_data.shape)
    show(scaled_data)
```
